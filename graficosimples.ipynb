```python
#importando a biblioteca
import matplotlib.pyplot as plt

#criando duas listas vazias para inserir os dados
x = []
y = []

#coletando as informações básicas com o usuário
titulo = input("Insira um nome para o gráfico: ")
nomex = input("Insira um nome para o eixo X: ")
nomey = input("Insira um nome para o eixo Y: ")
quantidade = int(input("Quantos dados terão para os eixos X e Y: "))

#criando um loop com as quantidades que o usuário escolheu
for i in range(quantidade):
  eixoX = int(input("Digite o dado para X: "))
  eixoY = int(input("Digite o dado para y: "))

  #adicionando os dados para as listas vazias
  x.append(eixoX)
  y.append(eixoY)

#crinado o gráfico
plt.plot(x, y, marker='o', linestyle='-', color='b')

#adicionando os títulos ao gráfico
plt.title(titulo)
plt.xlabel(nomex)
plt.ylabel(nomey)

#exibindo o gráfico
plt.show()
```
